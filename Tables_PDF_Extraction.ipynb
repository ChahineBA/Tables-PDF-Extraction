```python
import camelot  # Import the Camelot library for extracting tables from PDFs

# Extract tables from the specified PDF file
tables = camelot.read_pdf(
    r'/content/assurances_biat_2023.pdf',  # Path to the PDF file from which tables are to be extracted
    backend='poppler',                       # Use the Poppler backend for PDF parsing
    pages='3',                              # Extract tables from page 3 of the PDF
    flavor='stream',                        # Use the stream flavor for table extraction, suitable for tables without clear borders
    split_text=True                         # Split text across cells for better accuracy in extraction
)
```

```python
# Print the parsing report for the first table extracted
print(tables[0].parsing_report)
```

```python
# Access the DataFrame for the first extracted table
table = tables[0].df

# Display the DataFrame
table
```

```python
# Skip the first 3 rows and reset the index of the DataFrame
df = table.iloc[3:].reset_index(drop=True)

# Extract the first row as the new header
new_header = df.iloc[0]

# Drop the first row which was used as the new header
df = df[1:]

# Assign the new header to the DataFrame columns
df.columns = new_header

# Display the cleaned DataFrame
df
```

```python
import camelot
from tqdm import tqdm
import pandas as pd

# List of pages to extract tables from
pages = ["2", "3", "4", "5", "6", "7", "8", "27", "28", "29", "30", "31", "32", "33", "34"]

# Counter for naming sheets in the Excel file
p = 1

# Create an Excel writer object using OpenPyXL as the engine
with pd.ExcelWriter('tables.xlsx', engine='openpyxl') as excel_writer:
    # Iterate over each page specified in the pages list
    for page in tqdm(pages, desc="Extracting tables"):
        # Read tables from the current page of the PDF
        tables = camelot.read_pdf(
            '/content/assurances_biat_2023.pdf',  # Path to the PDF file
            backend='poppler',                       # Use Poppler backend for PDF processing
            pages=page,                              # Extract tables from the specified page
            flavor='stream',                        # Use the stream flavor for table extraction
            split_text=True                         # Split text across cells
        )

        # Print parsing report for debugging and verification
        print(tables[0].parsing_report)

        # Access the DataFrame for the first extracted table
        df = tables[0].df

        # Create a sheet name for the current table
        sheet_name = f'Table_{p}'

        # Write the DataFrame to the specified sheet in the Excel file
        df.to_excel(excel_writer, sheet_name=sheet_name, index=False)

        # Print a message indicating that the current table has been processed
        print(f"Done with table: {p}")

        # Increment the sheet counter
        p += 1
```
